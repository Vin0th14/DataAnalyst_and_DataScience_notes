<b>What is Variance Inflation Factor?</b><br/>
It's a metric or a score that tells us how well the variance captured by a particular column is explained by combination of other columns in the dataset.

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston

import statsmodels.api as sm
from sklearn.linear_model import LinearRegression

import warnings
warnings.filterwarnings('ignore')
```

```python
boston = load_boston()
print(boston.DESCR)
```

```python
X = boston["data"]
y = boston["target"]
columns = list(boston.feature_names)
```

```python
df = pd.DataFrame(X, columns=columns)
```

```python
df.head()
```

## Method 1

```python
vif_data = {
    "Column": [],
    "VIF": []
}
for index in range(len(columns)):
    x = df.loc[:, df.columns!=columns[index]]
    y = df.loc[:, df.columns==columns[index]]
    
    model = sm.OLS(y, x)
    results = model.fit()
    
    r_squared = results.rsquared
    vif = 1/(1-r_squared)
    
    vif_data["Column"].append(columns[index])
    vif_data["VIF"].append(vif)

vif_df = pd.DataFrame(vif_data).sort_values(by="VIF", ascending=False)
```

## Method 2

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif_df = pd.DataFrame()
vif_df["Column"] = columns
vif_df["VIF"] = [variance_inflation_factor(X, i) for i in range(len(columns))]
vif_df.sort_values(by="VIF", ascending=False)
```
