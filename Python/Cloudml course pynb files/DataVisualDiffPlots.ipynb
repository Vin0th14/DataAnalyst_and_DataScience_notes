# Data Visualization and Efficient Use of Different Plots

### 1. Why Data Visualization is needed

It is always better to explore each data set using multiple exploratory techniques and compare the results. This step aims to understand the dataset, identify the missing values & outliers if any using visual and quantitative methods to get a sense of the story it tells. It suggests the next logical steps, questions, or areas of research for your project.



There are two main objectives for using data visualization:

- A visual context enables viewers to more easily detect patterns, trend or correlations.
- Pictures can make it easier to communicate statistical results to an audience.

### "We, humans, remember the pictures more easily than readable text"

To communicate your data effectively, it is essential to use suitable types of plot, graph or chart.

Here we will offers a guideline for selecting the right plot type for your data.

## Visualization for Numerical Column

### Univariate Analysis

### Histogram

It basically creates bins in various ranges in values and plots it where we can visualize how values are distributed.

image.png

### Displot

Displot is also known as the second Histogram because it is a slight improvement version of the Histogram. Displot gives us a KDE(Kernel Density Estimation) over histogram which explains PDF(Probability Density Function) which means what is the probability of each value occurring in this column.

image.png

### Boxplot

Boxplot is a very interesting plot that basically plots a 5 number summary. i.e. Minimum and Maximum , Percentile , Median .

`Median` – Middle value in series after sorting.

`Percentile` – 25th , 50th and 75th percentile for given data distribution.

`Minimum and Maximum` – These are not minimum and maximum values, rather they describe the lower and upper boundary of standard deviation which is calculated using Interquartile range(IQR).

## Object Oriented Ways:

````python
fig=py.figure() # initiating figure

#set axes
axes1= fig.add_axes ([0.2,0.2,0.9,0.9]) #left,bottom,width,height
#plotting
axes1.plot(arr1, arrr2)
axes1.set_xlabel('Arr1')
axes1.set_xlabel('Arr2')
axes1.set_title('Arr1 vs Arr2')
````

````python
fig=py.figure()
#set axes
axes1= fig.add_axes ([0.5,0.1,0.9,0.4])
#plotting
axes1.plot(arr1, arrr2)
axes1.set_xlabel('Arr1')
axes1.set_xlabel('Arr2')
axes1.set_title('Arr1 vs Arr2')
````

````python
#Plot under plot
fig=py.figure()
axes1= fig.add_axes ([0.1,0.1,0.8,0.8]) #Left,bottom,width,height
axes1.plot(arr1, arrr2)
axes1.set_xlabel('Arr1')
axes1.set_ylabel('Arr2')
axes1.set_title('Arr1 vs Arr2')

axes2= fig.add_axes ([0.3,0.3,0.3,0.3]) #Left,bottom,width,height
axes2.plot(arr1, arrr2)
axes2.set_xlabel('Arr1')
axes2.set_ylabel('Arr2')
axes2.set_title('Arr1 vs Arr2')
````

````python
#Subplots
fig,axis=py.subplots(nrows=1, ncols=2)
axis[0].plot(arr1,arrr2)
axis[1].plot(arr1,arrr2)
````

## Figure_size and DPI:

````python
fig,axis=py.subplots(figsize=(10,5))
axes1= fig.add_axes ([0.3,0.5,0.3,0.3]) #Left,bottom,width,height
axes1.plot(arr1, arrr2)
axes1.set_xlabel('Arr1')
axes1.set_ylabel('Arr2')
axes1.set_title('Arr1 vs Arr2')
````

````python
font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'darkred','size':15}
font3={'family':'serif','color':'green','size':15}
````

````python
fig=py.figure()
axes1= fig.add_axes ([0.5,0.1,0.9,0.4]) #Left,bottom,width,height
axes1.plot(arr1, arrr2)
axes1.set_xlabel('Arr1',fontdict=font3)
axes1.set_ylabel('Arr2',fontdict=font2)
axes1.set_title('Arr1 vs Arr2',fontdict=font1)
````

````python
fig,axes=py.subplots()
axes.plot(arr1,arrr2,'r.',label='arr1 vs arrr2')
axes.plot(arrr2,arr1,'g--',label='arrr2 vs arr1')
axes.legend()
````
